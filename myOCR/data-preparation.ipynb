```python
from datasets import load_dataset, Image as HFImage
```

```python
ds = load_dataset("LULab/myOCR")
```

```python
def prepare_dataset(example):
    image, text = example['text'].split("\t")
    text = text.replace("_", " ")
    example['image'] = image
    example['text'] = text

    return example
```

```python
ds['train'] = ds['train'].map(prepare_dataset)
ds['validation'] = ds['validation'].map(prepare_dataset)
```

```python
ds = ds.cast_column('image', HFImage())
ds = ds.select_columns(['image', 'text'])
```
