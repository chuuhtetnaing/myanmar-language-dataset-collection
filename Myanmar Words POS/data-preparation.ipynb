```python
from datasets import Dataset
from pathlib import Path
```

# Read the data from text files

```python
# Download the text-files folder from https://github.com/myanmartools/myanmar-words
base_path = "./text-files"
base = Path(base_path)

data = {"POS Tag (English)": [], "POS Tag (Myanmar)": [], "Alphabet": [], "Word": []}
pos_mapping = {
    "နာမ်များ": "Nouns",
    "နာမ်စားများ": "Pronouns",
    "ကြိယာများ": "Verbs",
    "နာမဝိသေသနများ": "Adjectives",
    "ကြိယာဝိသေသနများ": "Adverbs",
    "ဝိဘတ်များ": "Postpositions",
    "သမ္ဗန္ဓများ": "Conjunctions",
    "ပစ္စည်းများ": "Particles",
    "အာမေဍိတ်များ": "Interjections"
}

# Sort folders first
for folder in sorted(base.iterdir()):
    if folder.is_dir():
        # Sort files within each folder
        for file in sorted(folder.glob('*.txt')):
            content = file.read_text(encoding='utf-8').strip().split("\n")
            folder_name = folder.name
            file_name = file.name.split("-")[0]

            data["POS Tag (Myanmar)"].extend([folder_name.replace("များ", "")] * len(content))
            data["POS Tag (English)"].extend([pos_mapping[folder_name]] * len(content))
            data["Word"].extend(content)
            data["Alphabet"].extend([file_name] * len(content))
```

# Read into HuggingFace Dataset and Push

```python
dataset = Dataset.from_dict(data)
dataset
```

```python
dataset.push_to_hub("chuuhtetnaing/myanmar-words-pos-dataset", private=True, token="")
```
